```python
# Import relevant libraries.
import tensorflow as tf
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Map data into arrays
s=np.asarray([1,0,0])
ve=np.asarray([0,1,0])
vi=np.asarray([0,0,1])
data['f5'] = data['f5'].map({'Iris-setosa': s, 'Iris-versicolor': ve,'Iris-virginica':vi})
```

```python
data = data.iloc[np.random.permutation(len(data))]
```

```python
data
```

```python
# Reset the index.
data=data.reset_index(drop=True)
```

```python
data
```

```python
# Choosing the training data.
x_input=data.loc[0:105,['f1','f2','f3','f4']]
temp=data['f5']
y_input=temp[0:106]

# Choosing the test data.
x_test=data.loc[106:149,['f1','f2','f3','f4']]
y_test=temp[106:150]
```

```python
# Placeholders and variables.
x=tf.placeholder(tf.float32,shape=[None,4])
y_=tf.placeholder(tf.float32,shape=[None, 3])

# Weight and bias
W=tf.Variable(tf.zeros([4,3]))
b=tf.Variable(tf.zeros([3]))
```

```python
# Softmax function.
y = tf.nn.softmax(tf.matmul(x, W) + b)
```

```python
# Loss function. Measuring the performance of Softmax.
cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y), reduction_indices=[1]))
```

```python
# Optimizer that implements Adam algorithm. (Adam is good for deep-learning task)
train_step = tf.train.AdamOptimizer(0.01).minimize(cross_entropy)
```

```python
# Calculating the model's accuracy.
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
#Session parameters
sess = tf.InteractiveSession()

#Initializing variables
init = tf.global_variables_initializer()
sess.run(init)

#Number of iterations
epoch = 2000
```

```python
for step in range(epoch): # Use xrange if you're using Python 2.
   _, c=sess.run([train_step,cross_entropy], feed_dict={x: x_input, y_:[t for t in y_input.as_matrix()]})
   if step%500==0 :
       print (c)
```

```python
# Random testing at Sn.130
a=data.loc[125,['f1','f2','f3','f4']]
b=a.values.reshape(1,4)
largest = sess.run(tf.argmax(y,1), feed_dict={x: b})[0]
if largest==0:
    print ("The flower is :Iris-setosa")
elif largest==1:
    print ("The flower is :Iris-versicolor")
else :
    print ("The flower is :Iris-virginica")
```
