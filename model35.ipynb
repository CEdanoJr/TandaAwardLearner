```python
#Importing the packages or libraries to be use.
import tensorflow as tf
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Map data into arrays. Also called One hot encoding.
ot15 = np.asarray([1,0,0])
ot20 = np.asarray([0,1,0])
wk = np.asarray([0,0,1])
data['f7'] = data['f7'].map({'OT15': ot15, 'OT20': ot20, 'WK': wk})
```

```python
data
```

```python
#Data shuffling. Shuffles the data to avoid biases.
data = data.iloc[np.random.permutation(len(data))]
```

```python
data
```

```python
#Reset the indexing.
data = data.reset_index(drop = True)
```

```python
data
```

```python
#Divide the dataset into training and test.

#Training set.
x_input = data.loc[0:305, ['f1','f2','f3','f4','f5','f6']]
temp = data['f7']
y_input = temp[0:306]

#Testing data.
x_test = data.loc[306:405, ['f1','f2','f3','f4','f5','f6']]
y_test = temp[306:406]
```

```python
#Placeholders & variables. 6 features as input and 3 awards as output.
x = tf.placeholder(tf.float32, shape = [None, 6])
y_ = tf.placeholder(tf.float32, shape = [None, 3])

#Weight and Biases. It is used for softmax regression (parameters).
W = tf.Variable(tf.zeros([6, 3]))
b = tf.Variable(tf.zeros([3]))
```

```python
#Softmax regression (model)
y = tf.nn.softmax(tf.matmul(x, W) + b)
```

```python
#Loss function
#Measuring the performance of our softmax regression model.
cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y), reduction_indices = [1]))
```

```python
#Optimizer that implements Adam algorithm. (Adam is good for deep-learning task)
train_step = tf.train.AdamOptimizer(0.01).minimize(cross_entropy)

#Calculating the model's accuracy
correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
#Session parameters
sess = tf.InteractiveSession()

#Initializing variables
init = tf.global_variables_initializer()
sess.run(init)

#Number of iterations
epoch = 2000
```

```python
for step in xrange(epoch):
   _, c=sess.run([train_step,cross_entropy], feed_dict={x: x_input, y_:[t for t in y_input.as_matrix()]})
   if step%500==0 :
       print c
```

```python
#Random testing @ 307-404 (the data we separate from training)
a = data.loc[402, ['f1','f2','f3','f4','f5','f6']]
b = a.values.reshape(1, 6)
largest = sess.run(tf.argmax(y, 1), feed_dict = {x: b})[0]
if largest == 0:
    print "Prediction: Overtime Rate OT15"
elif largest == 1:
    print "Prediction: Overtime Rate OT20"
else :
    print "Prediction: Overtime Rate Weekend Multiplier"
```

```python
#Calculate the accuracy of the model.
print sess.run(accuracy, feed_dict = {x: x_test, y_: [t for t in y_test.as_matrix()]})
```
